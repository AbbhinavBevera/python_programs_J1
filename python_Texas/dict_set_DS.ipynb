# dict
1. dict is collection of items enclosed by {}-item is key:value pair


```python
info={"name":"john","age":30,"city":"newyork"}
print(info)  
print(type(info))
```

```python
# read data adn update data from dict
info={"name":"john","age":30,"city":"newyork"}
print(info["name"])
print(info["age"])  
print(info)
```

```python
info={"name":"john","age":30,"city":"newyork"}
print(info)  
info["age"]=31
print(info)
```

```python
#methods or functions in dict
print(dir(dict))
```

```python
info={"name":"john","age":30,"city":"newyork"}
print(info)  
print(list(info.keys())) 
print(list(info.values()))          
print(list(info.items()))  
```

```python
#update()
info={"name":"john","age":30,"city":"newyork"}
print(info)
info.update({"age":32,"email":"john@example.com"})
print(info)
```

```python
#pop
info={"name":"john","age":30,"city":"newyork"}
print(info)
info.pop("city")
print(info)
```

```python
#methods or functons
print(dir(set))
```

```python
#add
a={1,2,3}
print(a)
a.add(4)
print(a)
```

```python
#difference
a={1,2,3,4}
b={3,4,5,6}
c=a.difference(b)
print(c)

```

```python
#union
a={1,2,3}
b={3,4,5}
c=a.union(b)
print(c)
```

```python
#intersection
a={1,2,3,4}
b={3,4,5,6}
c=a.intersection(b)
print(c)
```

```python
#discard
a={1,2,3,4}
print(a)
a.discard(3)
print(a)
```
