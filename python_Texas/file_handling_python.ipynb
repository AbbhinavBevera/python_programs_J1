# file handling in python
1. file is collected data stored in permanent storage
2. file handling is used to read or write data to a file
3. types of files
   a. text file
   b. binary file
4. binary file is used to store images , audio , video etc
5. text file is used to store plain text data       

# file operating modes in python
1. read mode (r)
2. write mode (w)
3. append mode (a)
4. read and write mode (r+)
5. binary mode (b)


```python
#write a program to create file using python script
create_file = open("profile.txt", "w")
print ("File created successfully")
create_file.close()
```

```python
#write a program to wrtie info to profile.txt file using python script
write_file = open("profile.txt", "w")
write_file.write("Name: Shiva\n")
write_file.write("Age: 25\n")
write_file.write("City: New York\n")
print ("Data written successfully to file")
write_file.close()
```
