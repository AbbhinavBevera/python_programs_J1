# itertion in python

1. iteration are loops to use perform a task multiple times.
2. there are two types of iteration in python
   - for loops
   - while loops

'''python-for loop

''' python-while loop
while condition:

```python
# write a program to print 1 to 100 numbers using loop
for num in range (1, 101):
    print(num)  
```

```python
# write a program to iterate characters in a string using for loop
name = "Texas"
for char in name:
    print(char) 
```

```python
# write a program to print even numbers between 500 to 600 numbers using loop
for n in range (500, 601):
    if(n%2==0):
     print(n,end=",")
```

```python
# write a program to print odd numbers between 500 to 600 numbers using loop
for n in range (500, 601):
    if(n%2!=0):
     print(n,end=",")
```

```python
# write a program to find the vowels from the string using for loop
name=str(input("Enter the string:"))
for char in name:
    if char in 'aeiouAEIOU':
        print(char,end=",")
```

```python
# write a program to find the consonants from the string using for loop
name=str(input("Enter the string:"))
for char in name:
    if char not in 'aeiouAEIOU':
        print(char,end=",")
```

```python
# write a program to iterate a list using for loop
names = ["Texas", "Python", "Django", "Flask"]
for name in names:
    print(name)
```

```python
# write a program to find the sum of all numbers from 1 to 100 using while loop
sum = 0
num = 1
while num <= 100:
    sum += num
    num += 1
print("The sum of all numbers from 1 to 100 is:", sum)
```

```python
# write a program to find sum of all even numbers between 1 to 50 using while loop
sum = 0
num = 1
while num <= 50:
    if num % 2 == 0:
        sum += num
    num += 1
print("The sum of all even numbers between 1 to 50 is:", sum)
```

```python
# write a program to find the sum of all odd numbers between 1 to 50 using while loop
sum = 0
num = 1
while num <= 50:
    if num % 2 != 0:
        sum += num
    num += 1
print("The sum of all odd numbers between 1 to 50 is:", sum)
```
