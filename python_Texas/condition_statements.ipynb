#conditional statements in python

1.pythin have four types of conditional statements
  - if statement
  - if...else statement
  - if...elif...else statement
  - nested if statement

'''python syntax --if
if condition:
    # code to be executed if condition is true
    '''

'''python syntax --if...else
if condition:
    # code to be executed if condition is true
else:
    # code to be executed if condition is false
    '''

'''python syntax --if...elif...else
if condition1:
    # code to be executed if condition1 is true
elif condition2:
    # code to be executed if condition2 is true
else:
    # code to be executed if both conditions are false
    '''    

```python
#write a python program to find the given number is odd.
number = int(input("Enter a number: "))
if number % 2 != 0:
    print("The number is odd.")
```

```python
#write a python program to find the given number is even or odd.
number = int(input("Enter a number: "))
if number % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
```

```python
#write a python program to find the given number is negative or positive or zero.
n=int(input("Enter a number: "))
if n>0:
    print("The number is positive.")
elif n<0:
    print("The number is negative.")
else:
    print("The number is zero.") 
```

```python
"""
write a program to check the grades based on marks
1. if marks in between 80-100 then print A
2. if marks in between 60-79 then print B
3. if marks in between 40-59 then print C
4. if marks in between 25-39 then print D
5. if marks less than 25 then print F
6. if marks are other than 0-100 then print invalid marks enter again
"""

marks=int(input("Enter your marks: "))
if 80 <= marks <= 100:          
    print("Grade A")
elif 60 <= marks <= 79:
    print("Grade B")
elif 40 <= marks <= 59:
    print("Grade C")
elif 25 <= marks <= 39:
    print("Grade D")
elif 0 <=marks < 25:
    print("Grade F")
else:
    print("Invalid marks")
```

```python
'''
write a program to find then even number wuth number of digits
#1. if number in between 1 to 9 then output is one digit even number
#2. if number in between 10 to 99 then output is two digit even number 
#3. if number in between 100 to 999 then output is three digit even number
#4. if number in between 1000 to 9999 then output is four digit even number
#5. if number is other than above then output is weired even number
'''
num=int(input("Enter a number: "))
if 1 <= num <= 9 and num % 2 == 0:
    print("One digit even number")
elif 10 <= num <= 99 and num % 2 == 0:
    print("Two digit even number")
elif 100 <= num <= 999 and num % 2 == 0:
    print("Three digit even number")
elif 1000 <= num <= 9999 and num % 2 == 0:
    print("Four digit even number")
else:
    print("Weired even number")
```
