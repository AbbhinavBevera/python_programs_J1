====conditional statements======
1.write a python program to check whether a number is positive,negative,zero
2.write a python program to check if a number is even or odd.
3.write a python program to find the greatest of two number using if-else
4.write a python program to find the smallest of there numbers using nested if statement if statements
5..write a python program to check if a year is a leap year or not

```python
n = int(input("please enter number:"))

if (n > 0):
    print(n, "is positive number")
elif (n < 0):
    print(n, "is negative number")
else:
    print(n, "is zero")
```

```python

num = int(input("please enter number:"))

if (num % 2 == 0):
    print(num, "is even number")
else:
    print(num, "is odd number")
```

```python

n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))

if n1 > n2:
    print(n1, "is the greatest number")
elif n2 > n1:
    print(n2, "is the greatest number")
else:
    print("Both numbers are equal")
```

```python
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))
n3 = int(input("Enter third number: "))

if n1 < n2:
    if n1 < n3:
        print(n1, "is the smallest number")
    else:
        print(n3, "is the smallest number")
else:
    if n2 < n3:
        print(n2, "is the smallest number")
    else:
        print(n3, "is the smallest number")
```

```python
year = int(input("Please enter a year: "))

if (year % 400 == 0):
    print(year, "is a leap year")
elif (year % 100 == 0):
    print(year, "is not a leap year")
elif (year % 4 == 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year") 
```

======iterations======
1.write a python program to print number from 1 to 10 using for loop.
2.write a python program to calculate the sum of first 20 natural numbers using a loop
3.write a python program to print all even numbers betwwen 1 to 50
4.write a python program to find the factorial of a number using a loop 
5.write a python program to count the number of digits in a given number using while loop.

```python
for num in range(1, 11):
    print(num)
```

```python
sum = 0
for num in range(1, 21):   # loop from 1 to 20
    sum += num
print("The sum of first 20 natural numbers is:", sum)
```

```python
for num in range(1, 51):
    if num % 2 == 0:
        print(num, end=" ")
```

```python
n = int(input("Enter a number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i

print("Factorial of", n, "is:", fact)
```

```python
num = int(input("Enter a number: "))
count = 0

if num == 0:
    count = 1
else:
    while num > 0:
        num //= 10
        count += 1

print("Number of digits:", count)
```
